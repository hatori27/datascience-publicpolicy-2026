## 2. Libraries: Importing the Toolboxes

In Python, we bring a specialized toolbox to our desk using the `import` keyword. 

For example, let us import the `pandas` library, which is the world's most popular toolbox for handling tabular datasets (like economic and demographic spreadsheets):

```python

```

**What does `as pd` mean?**
It is just a short nickname we give the toolbox. Instead of writing out "pandas" every time we want to grab a tool from it, we just write "pd". It saves typing!

````python
# Import the pandas toolbox
import pandas as pd

# Now the pandas toolbox is open and sitting on our desk!
print("Pandas toolbox is ready to use!")
````

## 3. Functions: Grabbing and Running a Tool

To use a tool, we need to do two things:
1.  **Grab it from the correct toolbox** (unless it is already built-in) using a dot `.`. For example, `pd.read_parquet` means: "Go to the `pd` toolbox and grab the `read_parquet` tool."
2.  **Turn it on** using parentheses `()`. Inside the parentheses, we feed the tool what it needs to do its job. These inputs are called **arguments**.

Let us look at some examples.

````python
# Example A: Using a built-in tool (len) to measure the number of letters in an organization's name
# No toolbox is needed because len is already on our workbench!
org_name = "North Atlantic Treaty Organization"
letter_count = len(org_name)
print("The number of characters (including spaces) in the NATO name is:")
print(letter_count)
````

## 4. Custom Functions: Building Your Own Tools

If Python or its libraries do not have the exact tool you need, you can build your own using the `def` keyword (which stands for "define").

Here is the anatomy of a custom tool:

```python
def tool_name(input_material):
    # 1. Take the input_material
    # 2. Do some work on it
    # 3. Hand back the finished product using 'return'
```

Let us build a simple tool that calculates the total cost of an imported good after applying a customs tariff rate.

````python
# Define our custom tariff calculator tool
# This tool accepts two inputs: 'import_price' and 'tariff_rate'
def calculate_tariff_cost(import_price, tariff_rate):
    # 1. LOCAL VARIABLES (Variables inside the function)
    # tariff_amount is a "local" variable. It is created inside the function's
    # sandbox and only exists while this function is running.
    tariff_amount = import_price * tariff_rate
    
    # total_cost is also a local variable. You cannot access tariff_amount
    # or total_cost from outside this function. They disappear when the function ends.
    total_cost = import_price + tariff_amount
    
    # 2. THE RETURN KEYWORD
    # return is how we "ship" the final result out of the function's sandbox
    # so that the outside world can receive and save it.
    return total_cost
````

````python
# Let us use our custom tariff calculator tool!
# 3. GLOBAL VARIABLES (Variables outside the function)
# raw_good_price and policy_tariff are defined on our main workshop workbench.
# They are "global" variables and are saved on our desk for us to use anytime.
raw_good_price = 500.00
policy_tariff = 0.12 # 12% tariff rate

# 4. SAVING THE VARIABLE
# We call calculate_tariff_cost and pass our global variables into it.
# IMPORTANT: To save the result of the function, we MUST assign it to a new
# variable on our workbench. Here, we catch the returned value and save it as 'final_import_cost'.
# If we didn't assign it (e.g. if we just ran calculate_tariff_cost(raw_good_price, policy_tariff)),
# the result would print or calculate, but would be lost forever!
final_import_cost = calculate_tariff_cost(raw_good_price, policy_tariff)

print("The total import cost of a $500 product under a 12% tariff is:")
print(final_import_cost)
````

### 4.1 The Metaphor: The Sandbox (Inside) vs. The Work Desk (Outside)

One of the most common points of confusion for new programmers is where variables are defined and saved. Why can't we access `total_cost` outside of our function?

Let's use our office workshop metaphor:

*   **The Work Desk (Global Scope / Outside the Function):**
    Your main Jupyter notebook workspace is your large physical work desk. Any variable you create directly on the desk (like `raw_good_price = 500.00`) sits there permanently until you clear the desk or close the notebook. Anyone in the office can see it and use it.
*   **The Sandbox (Local Scope / Inside the Function):**
    When Python runs a function, it temporarily builds a closed cardboard box (a sandbox) on your desk.
    1. It takes the ingredients you passed in (`import_price`, `tariff_rate`), puts them inside the box, and does the calculations.
    2. Any new variables created *inside* the box (like `tariff_amount` and `total_cost`) are temporary sticky notes stuck inside that box.
    3. As soon as the function hits `return`, it hands you the final product (`total_cost`) and **immediately crushes and throws away the cardboard box** along with all the sticky notes inside it.

This means:
*   **You cannot look inside the box after it's gone:** If you try to run `print(total_cost)` outside the function, Python will give you an error (`NameError: name 'total_cost' is not defined`).
*   **You must catch the returned product:** When the function hands back the final product, you must actively save it to a variable on your main desk (e.g., `final_import_cost = calculate_tariff_cost(...)`). If you don't assign it to a variable, the product drops to the floor and is lost!


````python
# DEMONSTRATION: Try running this cell to see what happens!

# This will work because final_import_cost was saved on our main work desk
print("Successfully saved variable from the desk:")
print(final_import_cost)

# This will fail and throw an error because tariff_amount only existed inside the function's sandbox!
try:
    print(tariff_amount)
except NameError as e:
    print("\nError as expected! Python says:")
    print(e)
````

## 6. Summary: How It All Fits Together

Here is the complete picture of your digital workshop:

1.  **Variables**: Labeled boxes you use to store your raw materials and finished products (`inflation_rate = 0.034`).
2.  **Libraries**: Specialized toolboxes full of advanced tools that you fetch when you need them (`import pandas as pd`).
3.  **Functions**: The individual tools you grab and turn on by adding parentheses `()` at the end, passing arguments inside them to get a result.
4.  **Custom Functions (`def`)**: Specialized custom tools and formulas that you design yourself to make your policy analysis faster, cleaner, and repeatable.

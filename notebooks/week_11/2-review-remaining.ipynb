## Week 5: Reshaping Data

Reshaping data is crucial when aligning datasets for visualization or regression analysis. 
- **Wide format**: Each variable has its own column. 
- **Long format**: Variables are stacked into a key-value format (tidy data).

```python
import pandas as pd

# Create dummy wide data
data_wide = pd.DataFrame({
    'country': ['USA', 'CAN', 'MEX'],
    'gdp_2021': [23000, 2000, 1300],
    'gdp_2022': [25000, 2100, 1400]
})
print("Wide Format DataFrame:")
print(data_wide)

# Reshape wide to long using melt
data_long = pd.melt(data_wide, id_vars=['country'], value_vars=['gdp_2021', 'gdp_2022'], var_name='year', value_name='gdp')
print("\nLong Format DataFrame (after melt):")
print(data_long)

# Reshape long back to wide using pivot
data_wide_back = data_long.pivot(index='country', columns='year', values='gdp').reset_index()
print("\nWide Format DataFrame (after pivot):")
print(data_wide_back)
```

## Week 9: NLP and AI LLMs

Processing textual features for policy analysis requires NLP toolkits.
- **spaCy NLP Pipeline**: Provides tokenization, lemmatization, and POS tagging out-of-the-box.
- **Transformer models**: Captures semantic embeddings to extract textual sentiment.

```python
import spacy

# Load small English model with safety exception handling
try:
    nlp = spacy.load("en_core_web_sm")
    doc = nlp("Data science for public policy helps make better decisions.")
    
    # Print token text and part-of-speech tags
    for token in doc:
        print(f"Token: {token.text:<15} POS: {token.pos_}")
except Exception as e:
    print(f"spaCy model not loaded: {e}. Please run 'python -m spacy download en_core_web_sm' in your terminal.")
```
